## Classification example using TensorFlow
### topics:
1. How to work with real dataset
2. tf.estimator
3. Categorical vs Continuous features
4. Changing tf model

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
%matplotlib inline
```

```python
diabetes = pd.read_csv('pima-indians-diabetes.csv')
diabetes.head()
```

```python
diabetes.columns
```

```python
cols_to_norm = ['Number_pregnant', 'Glucose_concentration', 'Blood_pressure', 'Triceps',
       'Insulin', 'BMI', 'Pedigree']
```

```python
#normalize columns, using pandas trick
diabetes[cols_to_norm] = diabetes[cols_to_norm].apply(lambda x: (x-x.min())/(x.max()-x.min()))
```

```python
diabetes.head()
```

```python
#continuous features
num_preg = tf.feature_column.numeric_column('Number_pregnant')
plasma_gluc = tf.feature_column.numeric_column('Glucose_concentration')
dias_press = tf.feature_column.numeric_column('Blood_pressure')
tricep = tf.feature_column.numeric_column('Triceps')
insulin = tf.feature_column.numeric_column('Insulin')
bmi = tf.feature_column.numeric_column('BMI')
diabetes_pedigree = tf.feature_column.numeric_column('Pedigree')
# can be treated as a continuous column
age = tf.feature_column.numeric_column('Age')
```

```python
#we know that there'll be only 4 groups
assigned_group = tf.feature_column.categorical_column_with_vocabulary_list('Group',['A','B','C','D'])
```

```python
#alternative
#assigned_group = tf.feature_column.categorical_column_with_hash_bucket('Group',hash_bucket_size=10)
```

```python
diabetes['Age'].hist(bins=20)
```

```python
age_bucketized = tf.feature_column.bucketized_column(age,boundaries=[20,30,40,50,60,70,80])
```

```python
#features columns
feat_cols =[num_preg,plasma_gluc,dias_press,tricep,insulin,bmi,diabetes_pedigree,assigned_group, age_bucketized]
```

```python
#train-test split
x_data = diabetes.drop('Class',axis=1)
x_data.head()
```

```python
labels = diabetes['Class']
#labels
```

```python
#train-test split
x_train, x_test, y_train, y_test = train_test_split(x_data, labels, test_size=0.3, random_state=101)
```

```python
#We're going to use pandas input function instead of numpy because our data is already in a pandas dataframe
input_func = tf.estimator.inputs.pandas_input_fn(x = x_train, y = y_train, batch_size=10, num_epochs=1000, shuffle=True)
```

```python
#create the Linear Classifier model built in Tensorflow
model = tf.estimator.LinearClassifier(feature_columns=feat_cols, n_classes=2)
```

```python
#training
model.train(input_fn=input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test, y=y_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
results = model.evaluate(input_fn=eval_input_func)
```

Same as before but we're not passing a y value to the estimator input because we're performing an evaluation

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
predictions = model.predict(pred_input_func)
my_pred = list(predictions)
my_pred
```

Let's change the classifier from a Linear Classifier to a Dense Neural Network classifier using 3 layers with 10 neurons each

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,10,10],feature_columns=feat_cols,n_classes=2)
```

Unfortunately we can't reuse the same input function, due to categorical columns

```python
embedded_group_col = tf.feature_column.embedding_column(assigned_group, dimension=4)
```

```python
feat_cols = [num_preg,plasma_gluc,dias_press,tricep,insulin,bmi,diabetes_pedigree, embedded_group_col, age_bucketized]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1,shuffle=True)
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,10,10],feature_columns=feat_cols,n_classes=2)
```

```python
dnn_model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test,y=y_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
dnn_model.evaluate(eval_input_func)
```

Compared to the previous model we can see that LinearClassifier gave a better result
